# Normal Cornea ☣

```python
# 1. Install the library
!pip install ezdxf

import numpy as np
import ezdxf
from google.colab import files

# Parameters
R = 7.10          # Apical radius of curvature (mm)
Q = -0.65        # Asphericity (prolate ellipsoid)
diameter = 12.0  # Width of the cornea (mm)
num_points = 100

# 2. Math: Generate the profile
# Solving the conicoid equation for z (depth)
# (1+Q)z^2 - 2Rz + x^2 = 0
a = 1 + Q
b = -2 * R
x_limit = diameter / 2
c = x_limit**2

# Calculate max depth (z) at the edge
z_max = (-b - np.sqrt(b**2 - 4*a*c)) / (2*a)

z_values = np.linspace(0, z_max, num_points)
# Calculate x (height from axis) for every z
x_values = np.sqrt(2 * z_values * R - (1 + Q) * z_values**2)

# 3. Create the DXF file
doc = ezdxf.new('R2010')
msp = doc.modelspace()

# Create the curve points (Z is horizontal, X is vertical)
points = list(zip(z_values, x_values))

# Add the Spline (the cornea surface)
msp.add_spline(points)

# Add closing lines to create a "pie slice" for easy revolving
# Line from the edge down to the center axis
msp.add_line((z_values[-1], x_values[-1]), (z_values[-1], 0))
# Line along the axis back to the start (origin)
msp.add_line((z_values[-1], 0), (0, 0))

# 4. Save and Download
filename = "steepcornea.2_profile.dxf"
doc.saveas(filename)
print(f"File {filename} created. Downloading now...")
files.download(filename)
```
